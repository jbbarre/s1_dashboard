```python
import pandas as pd
import hvplot.pandas
```

```python
antarctica = pd.read_json('antarctica_2023_12d.json')
antarctica.head()
```

```python
keys = antarctica.columns.tolist()
keys[1:]
```

```python
df = pd.DataFrame()
```

```python
df['slc']=antarctica['slc']/(2*antarctica['pairs'])*100 # *2: we hav two slc per pair
df['offmap.in']=antarctica['offmap.in']/antarctica['pairs']*100
df['offmap']=antarctica['offmap']/antarctica['pairs']*100
df['calib']=antarctica['calib']/antarctica['pairs']*100
df['interf']=antarctica['calib']/antarctica['pairs']*100
df['deramp']=antarctica['calib']/antarctica['pairs']*100
df['geo']=antarctica['calib']/antarctica['pairs']*100
df['figure']=antarctica['figure']/antarctica['pairs']*100
df
```

```python
df.tail(1).values
```
